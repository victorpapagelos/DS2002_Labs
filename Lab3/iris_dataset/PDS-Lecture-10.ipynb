### Trick for broadcasting when dimensions do not match

```python
b2.shape, d2.shape
```

```python
# We expand the dimensions of the array corresponding to the mismatch. 
# So in this case (2, 4) and (3, 4) agree on the last axis (both are 4)
# But the first dimension 2 != 3 and non of them is 1
# the solution:
# go to higher dimensional array breaking the mismatch.
# (2, ?, 4)
# (?, 3, 4)
# and we insert an axis with size 1 in the place of the mismatch
# (2, 1, 4)
# (1, 3, 4)

# Note: you could also do it the other way around:
# (?, 2, 4)
# (3, ?, 4)
# and insert the axis of size 1:
# (1, 2, 4)
# (3, 1, 4)
```

```python
b23 = np.expand_dims(b2, axis=-2)
b23
```

```python
d23 = np.expand_dims(d2, axis=0)
d23
```

```python
b23.shape, d23.shape
```

```python
(b23 - d2)
```
